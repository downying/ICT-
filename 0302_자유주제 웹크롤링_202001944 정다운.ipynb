# 2. 교보문고 - 베스트셀러 정보 가져오기

```python
url = 'http://www.kyobobook.co.kr/bestSellerNew/bestseller.laf'
html = requests.get(url)

#BeautifulSoup 객체로 변환
soup = BeautifulSoup(html.text, 'html.parser') 

# 웹 문서 전체 출력
# print(soup) 

rank = soup.select('div.badge_flag badge_light_gray')
print(rank)
for i in rank :
    num = i.select('span')
    print(num)
```

# 3. 예스24 월별 베스트셀러 - 1위 도서 정보 가져오기

```python
url = 'http://www.yes24.com/Product/Goods/117188090'
html = requests.get(url)

#BeautifulSoup 객체로 변환
soup = BeautifulSoup(html.text, 'html.parser') 
```

```python
# select_one(.class명)

name = soup.select_one('.gd_name').text
print("책 제목: " + name)
```

```python
# 개발자도구에서 태그를 'COPY' - 'COPY SELECTOR' 저자명 추출
tauth = '#yDetailTopWrap > div.topColRgt > div.gd_infoTop > span.gd_pubArea > span.gd_auth > a'
books = soup.select(tauth)
auth = books[0].text
print("저자명: " + auth)

# select_one(.class명)

auth = soup.select_one('span[class="gd_auth"]>a').text.strip()
print("저자명: " + auth)
```

```python
# # select_one(.class명)

pub = soup.select_one('.gd_pub').text
print("출판사명: " + pub)
```

```python
# # select_one(.class명)
date = soup.select_one('.gd_date').text
print("출간일: " + date)
```

```python
# # select_one(.class명)
price = soup.select_one('.yes_m').text
print("정가: " + price)
```

```python
# # select_one(.class명)
sale = soup.select_one('.nor_price').text
print("판매가: " + sale)
```

```python
url = 'http://www.yes24.com/Product/Goods/117188090'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser') 

name = soup.select_one('.gd_name').text
bname = "책 제목: " + name
print(bname)

auth = soup.select_one('span[class="gd_auth"]>a').text.strip()
bauth = "저자명: " + auth
print(bauth)

pub = soup.select_one('.gd_pub').text
bpub = "출판사명: " + pub
print(bpub)

date = soup.select_one('.gd_date').text
bdate = "출간일: " + date 
print(bdate)

price = soup.select_one('.yes_m').text
bprice = "정가: " + price
print(bprice)

sale = soup.select_one('.nor_price').text
bsale = "판매가: " + sale
print(bsale)
```
